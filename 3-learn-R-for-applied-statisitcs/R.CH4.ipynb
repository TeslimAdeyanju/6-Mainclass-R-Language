Sure, here is an example of how you can perform some basic statistical calculations on the `iris` dataset in R:



```r
# Load the iris dataset
data(iris)

# Calculate the mean of Sepal.Length
mean_sepal_length <- mean(iris$Sepal.Length)
print(paste("Mean of Sepal.Length: ", mean_sepal_length))

# Calculate the median of Sepal.Width
median_sepal_width <- median(iris$Sepal.Width)
print(paste("Median of Sepal.Width: ", median_sepal_width))

# Calculate the standard deviation of Petal.Length
sd_petal_length <- sd(iris$Petal.Length)
print(paste("Standard Deviation of Petal.Length: ", sd_petal_length))

# Calculate the variance of Petal.Width
var_petal_width <- var(iris$Petal.Width)
print(paste("Variance of Petal.Width: ", var_petal_width))

# Summary statistics for all numeric columns
summary_stats <- summary(iris[,1:4])
print(summary_stats)
```



This script calculates the mean of `Sepal.Length`, the median of `Sepal.Width`, the standard deviation of `Petal.Length`, and the variance of `Petal.Width`. It also prints a summary of statistics for all numeric columns in the `iris` dataset. The `summary()` function returns the minimum, 1st quartile, median, mean, 3rd quartile, and maximum for each column.
