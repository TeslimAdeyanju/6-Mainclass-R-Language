```r
# Load ggplot2 library
library(ggplot2)

# Create a sample data frame with some NA values
data <- data.frame(
  x = c(1, 2, 3, 4, NA, 6),
  y = c(2, 4, 1, NA, 5, 3)
)

# Print the data frame to show the NA values
print(data)

# Create a scatterplot using geom_point() with na.rm = TRUE
ggplot(data, aes(x = x, y = y)) +
  geom_point(na.rm = TRUE) +
  labs(title = "Scatterplot with na.rm = TRUE",
       x = "X-axis",
       y = "Y-axis")
```

```r
# Load necessary library
library(ggplot2)

# Define the data points
drives <- c(23, 78, 130, 147, 156, 177, 184, 213)

# Convert to data frame
drives_df <- data.frame(distance = drives)

# Create the histogram
ggplot(drives_df, aes(x = distance)) +
  geom_histogram(binwidth = 50, fill = "skyblue", color = "black") +
  scale_x_continuous(breaks = seq(0, 250, by = 50)) +
  scale_y_continuous(breaks = seq(0, 3, by = 1)) +
  labs(
    title = "Histogram of Luiza's Driving Distances",
    x = "Driving distance (meters)",
    y = "Number of drives"
  ) +
  theme_minimal()
```
